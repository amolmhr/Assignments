```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
# Replace 'language_code' with the column name containing languages in your DataFrame

top_10_languages = df['language_code'].value_counts().nlargest(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_languages.index, y=top_10_languages.values)
plt.title('Top 10 Languages for Books')
plt.xlabel('Language')
plt.ylabel('Number of Books')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
