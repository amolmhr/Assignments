<b> What is the role of the 'else' block in a try-except statement? Provide an example
scenario where it would be useful.</b>

In a try-except statement in Python (and many other programming languages), the `else` block is used to specify a block of code that should be executed if no exceptions are raised within the `try` block. It allows you to define code that should run only if the code within the `try` block executes successfully without any exceptions.

Here's the basic structure of a try-except-else statement:

```python
try:
    # Code that might raise an exception
except ExceptionType:
    # Code to handle the exception
else:
    # Code to execute if no exceptions are raised in the try block
```

Here's an example scenario where the `else` block is useful:

```python
def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Division by zero is not allowed")
    else:
        print(f"The result of {a} / {b} is {result}")



<b> 2. Can a try-except block be nested inside another try-except block? Explain with an
example</b>

Yes, a try-except block can be nested inside another try-except block. This is known as exception handling with nested try-except blocks. It allows you to handle exceptions at different levels of granularity, where the inner try-except block can catch and handle specific exceptions while the outer try-except block can handle more general exceptions or provide a broader level of error handling.

Here's an example to illustrate nested try-except blocks:

```python
def division_example(a, b, c):
    try:
        result = a / b  # Attempt division
        print(f"{a} / {b} = {result}")
        try:
            result = result / c  # Attempt another division
            print(f"{result} / {c} = {result}")
        except ZeroDivisionError:
            print("Inner division by zero is not allowed")
    except ZeroDivisionError:
        print("Outer division by zero is not allowed")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usages:
division_example(10, 2, 0)  # Outer division by zero is not allowed
division_example(10, 0, 5)  # Outer division by zero is not allowed
division_example(10, 2, 5)  # 10 / 2 = 5.0, 5.0 / 5 = 1.0
```

<b> 4. What are some common exceptions that are built-in to Python?</b>

4. `TypeError`: Raised when an operation is performed on an object of inappropriate type. For example, trying to concatenate a string and an integer will result in a `TypeError`.

5. `ValueError`: Raised when a function receives an argument of the correct type but an inappropriate value. For instance, trying to convert a string that doesn't represent a valid integer to an integer will raise a `ValueError`.

6. `ZeroDivisionError`: Raised when you attempt to divide a number by zero, which is mathematically undefined.

7. `IndexError`: Raised when you try to access an index that is out of range in a sequence, such as a list or a string.

8. `KeyError`: Raised when you try to access a dictionary key that doesn't exist.

9. `FileNotFoundError`: Raised when you try to open a file that doesn't exist.

10. `IOError`: Raised for various I/O-related errors, such as when reading from or writing to a file fails.

11. `ImportError`: Raised when an imported module cannot be found or loaded.

12. `AttributeError`: Raised when you try to access an attribute or method of an object that doesn't exist.

13. `RuntimeError`: A generic exception that can be used for various runtime errors that don't have a more specific exception class.

14. `MemoryError`: Raised when an operation runs out of memory.

15. `KeyboardInterrupt`: Raised when the user interrupts the program's execution, typically by pressing Ctrl+C.

These are just some of the common built-in exceptions in Python. Python's extensive standard library also provides many other exception classes to handle specific scenarios, and you can create your own custom exceptions when needed, as explained in a previous answer.

<b> 6. Explain the purpose of log levels in Python logging and provide examples of when

In Python logging, log levels are used to categorize and prioritize log messages based on their severity or importance. The `logging` module defines several standard log levels, which allow you to control the verbosity of log output and filter which messages get recorded. These log levels are ordered by increasing severity, from least severe to most severe:

1. **DEBUG**: This is the lowest log level. It is typically used for messages that are only relevant during debugging and development. Debug messages provide detailed information about the internal workings of the program. For example:

   ```python
   import logging

   logging.debug("Debugging information: variable x = %d", x)

   ```python
   import logging

3. **WARNING**: Warnings are used to indicate potential issues or situations that may require attention but don't necessarily lead to program failure. For instance, a warning might be issued when using deprecated features. Example:

   ```python
   import logging

   logging.warning("Deprecated feature 'xyz' is being used.")
   ```

4. **ERROR**: Error messages indicate that a specific error or problem occurred, but the program can typically continue executing. Error messages are used for issues that need to be addressed but don't necessarily terminate the application. For example:

<b> 7. What are log formatters in Python logging, and how can you customise the log
message format using formatters?</b>

In Python logging, log formatters are used to define the format of log messages before they are written to log outputs (e.g., files, console, etc.). Formatters allow you to customize the structure and content of log messages, making them more informative and easier to read. You can use predefined format options or create your own custom formats.

Here's how you can use log formatters in Python logging and customize the log message format:

1. **Using Predefined Formatters**:

Python's `logging` module provides several built-in log formatters that you can use out of the box. Some common predefined formatters include:

   - `logging.Formatter.defaultFormatter`: This formatter is used by default if you don't specify a custom formatter. It includes the log level, the log message, and the line number where the log record was created.

   - `logging.Formatter.simpleFormatter`: This formatter includes only the log message.

   - `logging.Formatter.basicFormatter`: This formatter includes the log level, the log message, and the time when the log record was created.

   To use one of these predefined formatters, you can set it when configuring your logging system:

   ```python
   import logging

   formatter = logging.Formatter.defaultFormatter  # Use the default formatter
   handler = logging.StreamHandler()  # Example: Log to console
   handler.setFormatter(formatter)  # Set the formatter for the handler
   logger = logging.getLogger('my_logger')
   logger.addHandler(handler)
   ```

2. **Creating Custom Formatters**:

   You can also create custom formatters to define the exact format you want for your log messages. Custom formatters are instances of the `logging.Formatter` class, and you can specify a custom format string using placeholders, such as `%s`, `%d`, and `%(levelname)s`, to include various log record attributes. Here's an example:

   ```python
   import logging

   formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
   handler = logging.StreamHandler()
   handler.setFormatter(formatter)
   logger = logging.getLogger('my_logger')
   logger.addHandler(handler)
   ```

   In this example, we create a custom formatter that includes the timestamp, log level, and log message in the log format.

3. **Format Placeholders**:

   Here are some commonly used placeholders you can include in your custom log format:

   - `%(asctime)s`: The timestamp when the log record was created.
   - `%(levelname)s`: The log level (e.g., DEBUG, INFO, WARNING, ERROR, CRITICAL).
   - `%(message)s`: The log message itself.
   - `%(name)s`: The logger's name.
   - `%(filename)s`: The filename where the log record was created.
   - `%(lineno)d`: The line number where the log record was created.
   - `%(funcName)s`: The name of the function where the log record was created.

   You can combine these placeholders and add additional text and formatting as needed to create your desired log message format.

Using log formatters and customizing log message formats is essential for tailoring your logs to your application's needs, making it easier to parse and analyze log data, and ensuring that the log messages contain relevant information for debugging, monitoring, and troubleshooting.

<b>8. How can you set up logging to capture log messages from multiple modules or
classes in a Python application?</b>

In Python, you can set up logging to capture log messages from multiple modules or classes in your application by creating a hierarchical logger hierarchy using the `logging` module. This allows you to organize and configure logging for different parts of your application independently. Here's how you can set up logging for multiple modules or classes:

1. **Create and Configure a Logger for Each Module or Class**:

   In each module or class that needs logging, create a logger using `logging.getLogger(__name__)`. By using `__name__` as the logger name, you ensure that each logger is associated with the module or class from which it's called. This way, you can configure and filter log messages for each module or class separately.

   ```python
   # Module1.py
   import logging

   logger = logging.getLogger(__name__)

   def some_function():
       logger.debug("Debug message from Module1")

   # Module2.py
   import logging

   logger = logging.getLogger(__name__)

   def another_function():
       logger.info("Info message from Module2")
   ```

2. **Configure Logging for Your Application**:

   In your main application or a central configuration file, set up the logging system by configuring log handlers, log levels, and formatters. You can configure different log handlers (e.g., file handler, console handler) and levels for different parts of your application.

   ```python
   import logging

   # Configure the root logger (for overall application settings)
   logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

   # Configure loggers for specific modules or classes
   logging.getLogger("Module1").setLevel(logging.DEBUG)
   logging.getLogger("Module2").setLevel(logging.INFO)
   ```

3. **Add Handlers and Formatters** (optional):

   You can add specific log handlers and formatters for each logger or at the root level to control where log messages go and how they are formatted. For example:

   ```python
   import logging

   root_logger = logging.getLogger()

   # Configure a console handler for the root logger
   console_handler = logging.StreamHandler()
   console_handler.setLevel(logging.DEBUG)
   formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
   console_handler.setFormatter(formatter)
   root_logger.addHandler(console_handler)
   ```

4. **Capture Log Messages from Multiple Modules or Classes**:

   Now, when you call functions or methods within your modules or classes, the log messages will be captured by the respective loggers you created. Each logger can have its own log level and handlers, allowing you to control the granularity of logging for different parts of your application.

   ```python
   # In your application code
   from Module1 import some_function
   from Module2 import another_function

   some_function()  # Logs a debug message from Module1
   another_function()  # Logs an info message from Module2
   ```

By setting up logging in this hierarchical manner, you can capture and configure log messages from multiple modules or classes independently. This approach helps you maintain clean and organized log output while allowing for different log levels and output destinations for various parts of your application.

   - **Print Statements**: Print statements are usually visible only to the developer or user who runs the application on their local machine. They are not suitable for monitoring a running application in production.

   - `level=logging.INFO` sets the log level to INFO, which ensures that the "Hello, World!" message is logged.
   - `filemode='a'` specifies that we want to append new log entries to the existing log file. If the file doesn't exist, it will be created.
   - `format='%(asctime)s - %(levelname)s - %(message)s'` defines the format of the log messages, including a timestamp, log level, and the log message itself.

3. We use `logging.info('Hello, World!')` to log the message "Hello, World!" with an INFO log level.

4. Optionally, we can use `logging.shutdown()` to close the logging system, but this is not strictly necessary for this simple program.
